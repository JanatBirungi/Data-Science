### Summary report and visualization of the WeRateDogs tweet dataset

In this report, I share my findings from analysing tweets from the WeRateDogs Twitter page. WeRateDogs shared a dump of its tweets to be analysed by Udacity students. 

The dataset I chose for my analysis contained tweets from when the page started up until 1st August 2017 because, beyond that date, there weren’t images for the various tweets. 

The dataset contained tweet information like tweet ids, dog stages(doggo, puppo, pupper and floofer), retweets count, favourite count(likes), image predictions, a link to the image on the Udacity servers and the rating of the dog. 

From my analysis, here are my key takeaways:

- WeRateDogs' Twitter page made its first tweet on 15th November 2015.

- Looking at the different stages of dogs, the most tweeted-about stage is the pupper stage, with about 211 tweets about this specific stage, the second most popular stage was the doggo stage with 75 tweets, puppo stage had 24 tweets and floofer had 8 tweets.

- The algorithm’s first prediction of the dog image in the tweet had a mean confidence of 59.5% and predicted a dog 74% of the time, in all tweets considered for the analysis. 

- The most retweets the page has received were 70,005 retweets

- The most likes this page received were 143,723 likes

- The post that had the most retweets also had the most likes. 

- The most popular post, the one with the most retweets and most likes, was of a dog in a swimming pool.
