3. Use the Tweepy library to query additional data via the Twitter API (tweet_json.txt)

```python
import tweepy
from tweepy import OAuthHandler
import json
from timeit import default_timer as timer

# Query Twitter API for each tweet in the Twitter archive and save JSON in a text file
# These are hidden to comply with Twitter's API terms and conditions
consumer_key = 'HIDDEN'
consumer_secret = 'HIDDEN'
access_token = 'HIDDEN'
access_secret = 'HIDDEN'

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit=True)

# NOTE TO STUDENT WITH MOBILE VERIFICATION ISSUES:
# df_1 is a DataFrame with the twitter_archive_enhanced.csv file. You may have to
# change line 17 to match the name of your DataFrame with twitter_archive_enhanced.csv
# NOTE TO REVIEWER: this student had mobile verification issues so the following
# Twitter API code was sent to this student from a Udacity instructor
# Tweet IDs for which to gather additional data via Twitter's API
tweet_ids = twitter_archive_data.tweet_id.values
len(tweet_ids)

# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
# read in text data line by line into python list
with open('tweet_json.txt') as file:
    status = []
    for line in file:
        status.append(json.loads(line))
```

```python
# converting the list into a pandas dataframe and pick only the tweet id, favorite count and retweet count
df_tweet_json = pd.DataFrame(status) 
df_tweets = df_tweet_json[["id","retweet_count","favorite_count"]]
df_tweets.head()
```

### Issue #7: Make the dog stages columns into a single column with the exact description of the stage

#### Define
Create on column for the dog stages doggo, pupper, puppo and doggo

#### Code

```python
# First convert all 'none' in dog stages to null so that these can be unpivoted

colms = ["doggo","floofer","pupper","puppo"]
original_tweets_images[colms] = original_tweets_images[colms].replace({'None':''})
twt_arc_trial = pd.melt(original_tweets_images, id_vars= ['tweet_id', 'jpg_url', 'img_num', 'p1', 'p1_conf', 'p1_dog', 'p2', 'p2_conf',  'p2_dog', 'p3','p3_conf', 'p3_dog','in_reply_to_status_id', 'in_reply_to_user_id', 'timestamp', 'source', 'text', 'retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp', 'expanded_urls', 'rating_numerator', 'rating_denominator', 'name'],
                                                        var_name='dog_stages', value_name='dog_stage_values')
twt_arc_trial = twt_arc_trial[twt_arc_trial.dog_stage_values != '']

original_tweets_images = pd.merge(original_tweets_images,twt_arc_trial[['tweet_id','dog_stages']], on='tweet_id', how='left') 
original_tweets_images = original_tweets_images.drop(["doggo","floofer","pupper","puppo","in_reply_to_status_id","in_reply_to_user_id","retweeted_status_id","retweeted_status_user_id"], axis=1)    
```

### Issue #8: Incorrect data types for the different columns

#### Define
Convert the columns listed below using astype to their appropriate data types
- The timestamp and retweeted_status_timestamp column to datetime data type 
- convert dog stages to category data type
- convert name,text columns into str data type

image prediction

- change the data type for p1,p2 and p3 the algorithm prediction columns to str

#### Code

```python
#to datetime
# patients_clean.birthdate = pd.to_datetime(patients_clean.birthdate)
original_tweets_images.timestamp = pd.to_datetime(original_tweets_images.timestamp)
original_tweets_images.retweeted_status_timestamp = pd.to_datetime(original_tweets_images.retweeted_status_timestamp)

# to str
original_tweets_images.text = original_tweets_images.text.astype(str)
original_tweets_images.name = original_tweets_images.name.astype(str)
original_tweets_images.p1 = original_tweets_images.p1.astype(str)
original_tweets_images.p2 = original_tweets_images.p2.astype(str)
original_tweets_images.p3 = original_tweets_images.p3.astype(str)

# to category
original_tweets_images.dog_stages = original_tweets_images.dog_stages.astype('category')
```
